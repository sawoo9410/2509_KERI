## 1. 분석에 필요한 패키지 불러오기

```python
import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import cv2

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import models, transforms

from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix, classification_report,
    accuracy_score, precision_score, recall_score,
    f1_score, roc_curve, roc_auc_score
)

# GPU 설정
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)
```

- PyTorch, torchvision과 OpenCV, PIL, sklearn 활용
- GPU 사용 시 연산 가속

## 2. DataFrame 생성: 정상 OK vs. 비정상 NG

```python
class AutoImageClassification:
    def __init__(self, base_dir):
        # train/test 경로
        self.train_df = self._create_df(os.path.join(base_dir, 'train'))
        self.test_df  = self._create_df(os.path.join(base_dir, 'test'))

    @staticmethod
    def _create_df(folder):
        ok = glob.glob(f"{folder}/OK/*.png")
        ng = glob.glob(f"{folder}/NG/*.png")
        print(f'{folder}: OK={len(ok)}, NG={len(ng)}')
        df_ok = pd.DataFrame({'path': ok, 'label': 1})
        df_ng = pd.DataFrame({'path': ng, 'label': 0})
        return shuffle(pd.concat([df_ok, df_ng], ignore_index=True))

# 예시
base_dir = './data/Chromate_Anomaly_detecton_datasets/resized'
aic = AutoImageClassification(base_dir)
aic.train_df.head()
```

- path, label 컬럼 확인
- .head() 로 상위 데이터 확인

## 3. 샘플 이미지 시각화

```python
# 정상 4개
fig, axs = plt.subplots(2,2, figsize=(6,6))
for ax, p in zip(axs.ravel(), aic.train_df[aic.train_df.label==1].path[:4]):
    img = cv2.imread(p)
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); ax.set_title('OK'); ax.axis('off')
plt.show()

# 비정상 4개
fig, axs = plt.subplots(2,2, figsize=(6,6))
for ax, p in zip(axs.ravel(), aic.train_df[aic.train_df.label==0].path[:4]):
    img = cv2.imread(p)
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); ax.set_title('NG'); ax.axis('off')
plt.show()
```

## 4. Dataset & DataLoader 준비

```python
# 4.1 이미지 전처리
train_tf = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.5,)*3, (0.5,)*3)
])
test_tf = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,)*3, (0.5,)*3)
])

# 4.2 커스텀 Dataset
class ChromateDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df.reset_index(drop=True)
        self.transform = transform
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        path = self.df.loc[idx, 'path']
        img  = Image.open(path).convert('RGB')
        if self.transform: img = self.transform(img)
        label = self.df.loc[idx, 'label']
        return img, label

train_ds = ChromateDataset(aic.train_df, transform=train_tf)
test_ds  = ChromateDataset(aic.test_df,  transform=test_tf)

desc = f"Train: {len(train_ds)}, Test: {len(test_ds)}"
print(desc)

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True,  num_workers=2)
test_loader  = DataLoader(test_ds,  batch_size=32, shuffle=False, num_workers=2)
```

- Resize, RandomHorizontalFlip, Normalize 사용
- Dataset → DataLoader로 배치 생성

## 5. 전이학습 모델 정의 (VGG16)

```python
# 5.1 사전학습 VGG16 불러오기
backbone = models.vgg16(pretrained=True)
# 마지막 FC 제외
features = backbone.features

# 5.2 classifier 추가\ 
nclass = 1  # 이진 분류
model = nn.Sequential(
    features,
    nn.AdaptiveAvgPool2d((1,1)),
    nn.Flatten(),
    nn.Linear(512, 500), nn.ReLU(), nn.Dropout(0.5),
    nn.Linear(500, nclass), nn.Sigmoid()
).to(device)
print(model)
```

- AdaptiveAvgPool2d((1,1)) → Flatten → Linear 구조
- 출력에 Sigmoid 사용해 이진 확률 반환

## 6. 학습 및 검증 루프 작성

```python
criterion = nn.BCELoss()  # Sigmoid + BCELoss
optimizer = optim.Adam(model.parameters(), lr=1e-4)

epochs = 10
history = {'train_loss':[], 'train_acc':[], 'val_loss':[], 'val_acc':[]}

for ep in range(1, epochs+1):
    # -- Training --
    model.train()
    tloss, tcorrect = 0, 0
    for imgs, labels in train_loader:
        imgs, labels = imgs.to(device), labels.float().unsqueeze(1).to(device)
        preds = model(imgs)
        loss = criterion(preds, labels)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        tloss += loss.item() * imgs.size(0)
        tcorrect += ((preds>0.5)==labels).sum().item()
    train_loss = tloss/len(train_ds)
    train_acc  = tcorrect/len(train_ds)

    # -- Validation --
    model.eval()
    vloss, vcorrect = 0, 0
    with torch.no_grad():
        for imgs, labels in test_loader:
            imgs, labels = imgs.to(device), labels.float().unsqueeze(1).to(device)
            preds = model(imgs)
            loss = criterion(preds, labels)
            vloss += loss.item() * imgs.size(0)
            vcorrect += ((preds>0.5)==labels).sum().item()
    val_loss = vloss/len(test_ds)
    val_acc  = vcorrect/len(test_ds)

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    print(f"Epoch {ep}/{epochs} "
          f"Train Loss:{train_loss:.4f} Acc:{train_acc:.4f} "
          f"Val Loss:{val_loss:.4f} Acc:{val_acc:.4f}")
```

```python
# 7.1 곡선 시각화
eps = range(1, epochs+1)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(eps, history['train_loss'], 'b-', label='Train Loss')
plt.plot(eps, history['val_loss'],   'r-', label='Val Loss')
plt.legend(); plt.title('Loss'); plt.grid()

plt.subplot(1,2,2)
plt.plot(eps, history['train_acc'], 'b-', label='Train Acc')
plt.plot(eps, history['val_acc'],   'r-', label='Val Acc')
plt.legend(); plt.title('Accuracy'); plt.grid()
plt.show()

# 7.2 Confusion Matrix & ROC
def eval_metrics(loader):
    all_preds, all_labels, all_probs = [], [], []
    model.eval()
    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device)
            probs = model(imgs).cpu().numpy().flatten()
            preds = (probs>0.5).astype(int)
            all_preds.extend(preds)
            all_labels.extend(labels.numpy())
            all_probs.extend(probs)
    return np.array(all_labels), np.array(all_preds), np.array(all_probs)

y_true, y_pred, y_proba = eval_metrics(test_loader)
print('Classification Report')
print(classification_report(y_true, y_pred, target_names=['NG','OK']))

fpr, tpr, _ = roc_curve(y_true, y_proba)
auc = roc_auc_score(y_true, y_proba)
plt.plot(fpr, tpr, label=f'AUC={auc:.4f}')
plt.plot([0,1],[0,1],'k--'); plt.xlabel('FPR'); plt.ylabel('TPR'); plt.legend(); plt.show()
```

### 6.1 해석

CNN Model의 평가를 위해 테스트 데이터셋을 사용하여 Predict 함수를 통해 예측하고,<br>
테스트 데이터셋 target 데이터와 비교하여 정확도가 100%로 나타난다.<br>

입력 데이터로 **이미지 데이터를 입력하였을 때 나오는 예측값**을 통해 **이미지 데이터가 어떤 품질 결과가 나타나는지 확인**할 수 있다.<br>
이를 통해, **육안검사를 통해 진행하던 품질검사를 자동으로 진행하는 공정으로 진행할 수 있다.**
