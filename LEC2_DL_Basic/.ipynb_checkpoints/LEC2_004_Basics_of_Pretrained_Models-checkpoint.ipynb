# 1. 학습에 필요한 라이브러리 불러오기

```python
import torch
from torchvision import models, transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

# 2. Pre-trained Model 불러오기

```python
# 1. 사전학습된 VGG16 모델 불러오기 (ImageNet weight, eval 모드)
model = models.vgg16(pretrained=True)
model.eval()
```

```python
# 모델 구조 출력
print(model)
```

```python
# 2. 이미지 불러오기 및 전처리
img_path = './data/banana.jpg'
img = Image.open(img_path).convert('RGB')    # 이미지 읽기
```

```python
# torchvision 전처리 (VGG16용 ImageNet 기준)
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),                # 이미지 크기 조정
    transforms.ToTensor(),                        # tensor 변환 (0-1)
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # 사전학습 가중치 기준 정규화
])

input_tensor = preprocess(img)                     # (3, 224, 224)
input_batch = input_tensor.unsqueeze(0)            # (1, 3, 224, 224)

print("image data shape:", np.array(img).shape)
plt.imshow(np.asarray(img))
plt.axis('off')
plt.show()

print("input data tensor shape:", input_batch.shape)
```

```python
# 3. 예측 수행
with torch.no_grad():
    outputs = model(input_batch)
    _, indices = outputs.topk(3)          # 상위 3개 예측 결과 인덱스
```

```python
# 4. ImageNet 클래스 라벨 매핑 불러오기
# 사전에 제공되는 labels 파일(local 혹은 URL/json 이용)
import urllib.request
import json

LABELS_URL = "https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt"
labels_path = "imagenet_classes.txt"
urllib.request.urlretrieve(LABELS_URL, labels_path)

with open(labels_path, "r") as f:
    categories = [s.strip() for s in f.readlines()]
```

# 3. 모델 예측 결과

```python
# 5. Top-3 예측결과 출력
for i in indices[0]:
    print(f"Predicted: {categories[i]}, Prob: {torch.softmax(outputs, 1)[0][i].item():.3f}")
```
