## 모델-프리 강화학습 (Model-Free Reinforcement Learning)

모델-프리 학습이 필요한 이유:
1. 현실 세계의 복잡성
- 로봇 제어: 환경의 전이 확률을 사전에 알기 어려움
- 금융 거래: 시장의 높은 변동성과 예측 불가능성
- 게임 AI: 상대방의 전략과 환경 변화에 적응 필요

2. 모델 획득의 어려움
- 전이 확률과 보상 함수를 직접 측정하기 어려움
- 상태/행동 공간이 클 때 모든 조합 저장 불가능
- 환경이 시간에 따라 변하는 경우 고정 모델 한계

3. 모델-프리 접근법의 장점
- 경험 기반 학습: 실제 상호작용만으로 학습
- 적응성: 환경 변화에 자동으로 적응
- 구현 단순성: 복잡한 모델 구축 불필요

### 환경 설정: Cliff Walking 문제 

```python
class CliffWalkingEnv:
    """
    Cliff Walking 환경 (개선된 버전)
    - 4x12 격자에서 좌하단(S)에서 우하단(G)으로 이동
    - 하단 절벽 구역을 지나가면 -100 패널티와 함께 시작점 복귀
    - 일반 이동은 -1, 목표 도달은 0
    """
    
    def __init__(self):
        self.rows = 4
        self.cols = 12
        self.n_states = self.rows * self.cols
        self.n_actions = 4  # 상하좌우
        
        # 특수 위치 정의
        self.start_state = self.coord_to_state(3, 0)  # 좌하단
        self.goal_state = self.coord_to_state(3, 11)  # 우하단
        self.cliff_states = [self.coord_to_state(3, i) for i in range(1, 11)]
        
        # 행동 정의: 0=상, 1=하, 2=좌, 3=우
        self.actions = ['↑', '↓', '←', '→']
        self.action_effects = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        
        self.reset()
    
    def coord_to_state(self, row, col):
        """좌표를 상태 번호로 변환"""
        return row * self.cols + col
    
    def state_to_coord(self, state):
        """상태 번호를 좌표로 변환"""
        return (state // self.cols, state % self.cols)
    
    def reset(self):
        """환경 초기화"""
        self.current_state = self.start_state
        return self.current_state
    
    def step(self, action):
        """행동 실행"""
        row, col = self.state_to_coord(self.current_state)
        d_row, d_col = self.action_effects[action]
        
        # 새로운 위치 계산 (경계 처리)
        new_row = max(0, min(self.rows - 1, row + d_row))
        new_col = max(0, min(self.cols - 1, col + d_col))
        next_state = self.coord_to_state(new_row, new_col)
        
        # 보상 계산
        if next_state == self.goal_state:
            reward = 0
            done = True
        elif next_state in self.cliff_states:
            reward = -100
            next_state = self.start_state  # 시작점 복귀
            done = False
        else:
            reward = -1
            done = False
        
        self.current_state = next_state
        return next_state, reward, done
    
    def get_valid_actions(self, state):
        """해당 상태에서 가능한 행동들"""
        return list(range(self.n_actions))
    
    def is_terminal(self, state):
        """종료 상태인지 확인"""
        return state == self.goal_state

# 환경 초기화
env = CliffWalkingEnv()

print("Cliff Walking 환경 설정:")
print(f"• 격자 크기: {env.rows}x{env.cols}")
print(f"• 시작 상태: {env.start_state} (좌하단)")
print(f"• 목표 상태: {env.goal_state} (우하단)")
print(f"• 절벽 상태: {env.cliff_states}")
print(f"• 행동: {env.actions}")
```

```python
# 환경 시각화 함수
def visualize_cliff_world(values=None, policy=None, title="Cliff Walking World", ax=None):
    """Cliff Walking 환경 시각화"""
    if ax is None:
        fig, ax = plt.subplots(figsize=(14, 6))
        show_plot = True
    else:
        show_plot = False
    
    # 기본 격자 그리기
    for i in range(env.rows + 1):
        ax.axhline(i - 0.5, color='black', linewidth=1)
    for j in range(env.cols + 1):
        ax.axvline(j - 0.5, color='black', linewidth=1)
    
    # 각 셀 채우기
    for state in range(env.n_states):
        row, col = env.state_to_coord(state)
        display_row = env.rows - 1 - row  # y축 뒤집기 (시각적 개선)
        
        # 절벽은 빨간색
        if state in env.cliff_states:
            ax.add_patch(Rectangle((col-0.4, display_row-0.4), 0.8, 0.8, 
                                 facecolor='red', alpha=0.7))
            ax.text(col, display_row, 'CLIFF', ha='center', va='center',
                   fontsize=8, fontweight='bold', color='white')
        
        # 시작점은 초록색
        elif state == env.start_state:
            ax.add_patch(Rectangle((col-0.4, display_row-0.4), 0.8, 0.8,
                                 facecolor='green', alpha=0.7))
            ax.text(col, display_row, 'S', ha='center', va='center',
                   fontsize=16, fontweight='bold', color='white')
        
        # 목표점은 파란색
        elif state == env.goal_state:
            ax.add_patch(Rectangle((col-0.4, display_row-0.4), 0.8, 0.8,
                                 facecolor='blue', alpha=0.7))
            ax.text(col, display_row, 'G', ha='center', va='center',
                   fontsize=16, fontweight='bold', color='white')
        
        # 일반 상태
        else:
            # 가치 함수가 주어진 경우
            if values is not None and state in values:
                value = values[state]
                # 값에 따른 색상 설정 (개선된 색상 스케일)
                if value < -50:
                    color = 'darkred'
                elif value < -30:
                    color = 'red'
                elif value < -20:
                    color = 'orange'
                elif value < -10:
                    color = 'yellow'
                else:
                    color = 'lightblue'
                
                ax.add_patch(Rectangle((col-0.4, display_row-0.4), 0.8, 0.8,
                                     facecolor=color, alpha=0.6))
                ax.text(col, display_row, f'{value:.0f}', ha='center', va='center',
                       fontsize=9, fontweight='bold')
            
            # 정책이 주어진 경우 화살표 표시
            if policy is not None and state in policy:
                best_action = policy[state]
                arrow_props = {
                    0: (0, 0.2),    # 상 ↑
                    1: (0, -0.2),   # 하 ↓
                    2: (-0.2, 0),   # 좌 ←
                    3: (0.2, 0)     # 우 →
                }
                
                if best_action in arrow_props:
                    dx, dy = arrow_props[best_action]
                    ax.arrow(col, display_row, dx, dy, 
                            head_width=0.08, head_length=0.08,
                            fc='black', ec='black', linewidth=2)
    
    ax.set_xlim(-0.5, env.cols - 0.5)
    ax.set_ylim(-0.5, env.rows - 0.5)
    ax.set_aspect('equal')
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel('Column')
    ax.set_ylabel('Row')
    
    # 좌표 레이블 추가
    ax.set_xticks(range(env.cols))
    ax.set_yticks(range(env.rows))
    
    if show_plot:
        plt.tight_layout()
        plt.show()

# 초기 환경 시각화
visualize_cliff_world(title="Cliff Walking Env. (S: Start, G: Goal, CLIFF: Cliff)")
```

## Chapter 5: 몬테카를로 학습 (Monte Carlo Methods)

몬테카를로 방법의 핵심 아이디어:
- 복잡한 문제를 무작위 샘플링으로 근사해 해결
- 강화학습에서: 실제 에피소드 경험으로 가치 함수 추정
- 특징: 에피소드 완료 후 업데이트 (편향 없는 추정)

### 5-1. Monte Carlo Prediction (정책 평가)

```python
def mc_prediction(env, policy, num_episodes=1000, gamma=1.0, first_visit=True):
    """
    몬테카를로 예측: 주어진 정책의 상태 가치 함수 추정 (개선된 버전)
    """
    V = defaultdict(float)
    returns = defaultdict(list)
    returns_history = []
    visit_counts = defaultdict(int)
    
    print(f"몬테카를로 예측 시작 ({num_episodes}개 에피소드)")
    print(f"방법: {'First-Visit' if first_visit else 'Every-Visit'} MC")
    
    for episode in range(num_episodes):
        # 에피소드 실행
        states = []
        actions = []
        rewards = []
        
        state = env.reset()
        done = False
        step_count = 0
        max_steps = 200  # 무한 루프 방지
        
        # 에피소드 궤적 생성
        while not done and step_count < max_steps:
            # 정책에 따라 행동 선택
            if state in policy:
                action_probs = policy[state]
                action = np.random.choice(len(action_probs), p=action_probs)
            else:
                action = np.random.choice(env.n_actions)
            
            states.append(state)
            actions.append(action)
            
            next_state, reward, done = env.step(action)
            rewards.append(reward)
            
            state = next_state
            step_count += 1
        
        # 리턴(G_t) 계산
        G = 0
        visited_states = set() if first_visit else None
        
        # 역순으로 리턴 계산
        for t in reversed(range(len(states))):
            G = gamma * G + rewards[t]
            state_t = states[t]
            
            # First-Visit: 처음 방문한 상태만
            if first_visit:
                if state_t not in visited_states:
                    returns[state_t].append(G)
                    visit_counts[state_t] += 1
                    visited_states.add(state_t)
            # Every-Visit: 모든 방문
            else:
                returns[state_t].append(G)
                visit_counts[state_t] += 1
        
        # 가치 함수 업데이트 (모든 상태에 대해)
        for state_t in returns:
            V[state_t] = np.mean(returns[state_t])
        
        # 주기적으로 진행상황 출력
        if (episode + 1) % 400 == 0:
            avg_return = np.mean([np.mean(rets) for rets in returns.values()]) if returns else 0
            print(f"에피소드 {episode + 1}: 평균 상태 가치 = {avg_return:.2f}, 방문한 상태 수 = {len(V)}")
        
        # 기록 저장 (일부 에피소드만)
        if episode % 100 == 0:
            returns_history.append(dict(V))
    
    return dict(V), returns_history, dict(visit_counts)

# 간단한 랜덤 정책 생성 (개선된 버전)
def create_random_policy(env):
    """균등 랜덤 정책 생성"""
    policy = {}
    for state in range(env.n_states):
        if state not in env.cliff_states and state != env.goal_state:
            policy[state] = np.ones(env.n_actions) / env.n_actions
    return policy

# 몬테카를로 예측 실행
random_policy = create_random_policy(env)
V_mc, mc_history, mc_visits = mc_prediction(env, random_policy, num_episodes=3000, first_visit=True)

print(f"\n몬테카를로 예측 결과:")
print(f"학습된 상태 수: {len(V_mc)}")

# 주요 상태들의 가치 출력
key_states = [env.start_state, env.coord_to_state(2, 5), env.coord_to_state(1, 10)]
for state in key_states:
    row, col = env.state_to_coord(state)
    visits = mc_visits.get(state, 0)
    print(f"상태 ({row},{col}): 가치={V_mc.get(state, 0):.2f}, 방문횟수={visits}")

# 결과 시각화
visualize_cliff_world(V_mc, title="MC Prediction: State Value of Random Policy")
```

### 5-2. Monte Carlo Control (정책 개선)

```python
def mc_control(env, num_episodes=3000, gamma=1.0, epsilon=0.1, epsilon_decay=0.995):
    """
    몬테카를로 제어: 최적 정책 학습 (개선된 버전)
    """
    Q = defaultdict(lambda: np.zeros(env.n_actions))
    returns = defaultdict(list)
    policy = {}
    episode_rewards = []
    episode_lengths = []
    
    print(f"몬테카를로 제어 시작 ({num_episodes}개 에피소드)")
    print(f"초기 탐험률: {epsilon}, 감소율: {epsilon_decay}")
    
    for episode in range(num_episodes):
        # 에피소드 실행
        states = []
        actions = []
        rewards = []
        
        state = env.reset()
        done = False
        step_count = 0
        max_steps = 200
        episode_reward = 0
        
        # 현재 에피소드의 epsilon 값
        current_epsilon = epsilon * (epsilon_decay ** episode)
        
        # 에피소드 궤적 생성
        while not done and step_count < max_steps:
            # epsilon-greedy 정책으로 행동 선택
            if np.random.random() < current_epsilon:
                action = np.random.choice(env.n_actions)
            else:
                action = np.argmax(Q[state])
            
            states.append(state)
            actions.append(action)
            
            next_state, reward, done = env.step(action)
            rewards.append(reward)
            episode_reward += reward
            
            state = next_state
            step_count += 1
        
        episode_rewards.append(episode_reward)
        episode_lengths.append(step_count)
        
        # 리턴 계산 및 Q 함수 업데이트
        G = 0
        visited_pairs = set()
        
        for t in reversed(range(len(states))):
            G = gamma * G + rewards[t]
            state_action = (states[t], actions[t])
            
            if state_action not in visited_pairs:
                returns[state_action].append(G)
                Q[states[t]][actions[t]] = np.mean(returns[state_action])
                visited_pairs.add(state_action)
        
        # 정책 업데이트 (epsilon-greedy 정책)
        for state in Q:
            best_action = np.argmax(Q[state])
            action_probs = np.ones(env.n_actions) * current_epsilon / env.n_actions
            action_probs[best_action] += (1.0 - current_epsilon)
            policy[state] = action_probs
        
        # 주기적으로 진행상황 출력
        if (episode + 1) % 600 == 0:
            recent_rewards = episode_rewards[-100:]
            recent_lengths = episode_lengths[-100:]
            avg_reward = np.mean(recent_rewards)
            avg_length = np.mean(recent_lengths)
            print(f"에피소드 {episode + 1}: 평균 보상 = {avg_reward:.2f}, 평균 길이 = {avg_length:.1f}, ε = {current_epsilon:.3f}")
    
    # 최종 그리디 정책 추출
    final_policy = {}
    for state in Q:
        final_policy[state] = np.argmax(Q[state])
    
    return final_policy, dict(Q), episode_rewards, episode_lengths

# 몬테카를로 제어 실행
mc_policy, Q_mc, mc_rewards, mc_lengths = mc_control(env, num_episodes=6000, epsilon=0.2)

print(f"\n몬테카를로 제어 결과:")
print(f"학습된 상태-행동 쌍 수: {len(Q_mc)}")
```

```python
# 학습 곡선 시각화 (개선된 버전)
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
window_size = 100
if len(mc_rewards) >= window_size:
    moving_avg = []
    for i in range(window_size, len(mc_rewards)):
        moving_avg.append(np.mean(mc_rewards[i-window_size:i]))
    
    plt.plot(range(window_size, len(mc_rewards)), moving_avg, 'b-', linewidth=2)
    plt.xlabel('Episode')
    plt.ylabel('Average Reward')
    plt.title('MC Control Learning Curve\n(100 Episode move-average)')
    plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
if len(mc_lengths) >= window_size:
    moving_avg_length = []
    for i in range(window_size, len(mc_lengths)):
        moving_avg_length.append(np.mean(mc_lengths[i-window_size:i]))
    
    plt.plot(range(window_size, len(mc_lengths)), moving_avg_length, 'g-', linewidth=2)
    plt.xlabel('Episode')
    plt.ylabel('Average Episode length')
    plt.title('Changing the length of an episode\n(100 Episode move-average)')
    plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
if len(mc_rewards) >= 1000:
    recent_rewards = mc_rewards[-1000:]
    plt.hist(recent_rewards, bins=30, alpha=0.7, color='green', edgecolor='black')
    plt.xlabel('Episode Reward')
    plt.ylabel('Frequancy')
    plt.title('Recently 1000 Episode\nReward Distribution')
    plt.axvline(np.mean(recent_rewards), color='red', linestyle='--', 
                linewidth=2, label=f'평균: {np.mean(recent_rewards):.1f}')
    plt.legend()

plt.tight_layout()
plt.show()

# 학습된 정책 시각화
visualize_cliff_world(policy=mc_policy, title="몬테카를로 제어: 학습된 최적 정책")
```

## Chapter 6: 시간차 학습 (Temporal Difference Learning)

시간차 학습의 핵심 아이디어:
- 에피소드 완료를 기다리지 않고 매 스텝마다 업데이트
- 부트스트래핑: 현재 추정치로 추정치를 개선
- 실시간 학습 가능, 연속적인 환경에 적합

### 6-1. TD(0) Prediction (시간차 예측)

```python
def td_prediction(env, policy, num_episodes=1000, alpha=0.1, gamma=1.0):
    """
    TD(0) 예측: 주어진 정책의 상태 가치 함수 추정 (개선된 버전)
    """
    V = defaultdict(float)
    td_errors_history = []
    visit_counts = defaultdict(int)
    
    print(f"TD(0) 예측 시작 ({num_episodes}개 에피소드)")
    print(f"학습률: {alpha}, 할인율: {gamma}")
    
    for episode in range(num_episodes):
        state = env.reset()
        done = False
        step_count = 0
        max_steps = 200
        episode_td_errors = []
        
        while not done and step_count < max_steps:
            visit_counts[state] += 1
            
            # 정책에 따라 행동 선택
            if state in policy:
                action_probs = policy[state]
                action = np.random.choice(len(action_probs), p=action_probs)
            else:
                action = np.random.choice(env.n_actions)
            
            next_state, reward, done = env.step(action)
            
            # TD 오차 계산 및 가치 함수 업데이트
            if not done:
                td_target = reward + gamma * V[next_state]
            else:
                td_target = reward
            
            td_error = td_target - V[state]
            episode_td_errors.append(abs(td_error))
            
            # 가치 함수 업데이트
            V[state] += alpha * td_error
            
            state = next_state
            step_count += 1
        
        # TD 오차 기록
        if episode_td_errors:
            td_errors_history.append(np.mean(episode_td_errors))
        
        # 주기적으로 진행상황 출력
        if (episode + 1) % 400 == 0:
            avg_td_error = np.mean(td_errors_history[-50:]) if len(td_errors_history) >= 50 else 0
            print(f"에피소드 {episode + 1}: 평균 TD 오차 = {avg_td_error:.4f}, 학습한 상태 수 = {len(V)}")
    
    return dict(V), td_errors_history, dict(visit_counts)

# TD(0) 예측 실행
V_td, td_errors, td_visits = td_prediction(env, random_policy, num_episodes=3000, alpha=0.1)

print(f"\nTD(0) 예측 결과:")
print(f"학습된 상태 수: {len(V_td)}")
```

### Chapter 7: SARSA와 Q-Learning

TD Control 방법:
- SARSA: On-policy 제어 (실제 따르는 정책으로 학습)
- Q-Learning: Off-policy 제어 (최적 정책을 가정하고 학습)
- 둘 다 행동 가치 함수 Q(s,a)를 학습하여 정책 개선

```python
# 7-1. SARSA (State-Action-Reward-State-Action) - 개선된 버전
print("\n7-1. SARSA Algorithm")
print("-" * 30)

def sarsa(env, num_episodes=3000, alpha=0.1, gamma=1.0, epsilon=0.1, epsilon_decay=0.995):
    """
    SARSA 알고리즘: On-policy TD 제어 (개선된 버전)
    """
    Q = defaultdict(lambda: np.zeros(env.n_actions))
    episode_rewards = []
    episode_lengths = []
    q_updates = []
    
    print(f"SARSA 학습 시작 ({num_episodes}개 에피소드)")
    print(f"학습률: {alpha}, 할인율: {gamma}, 초기 탐험률: {epsilon}")
    
    for episode in range(num_episodes):
        state = env.reset()
        
        # 현재 에피소드의 epsilon 값
        current_epsilon = epsilon * (epsilon_decay ** episode)
        
        # epsilon-greedy로 첫 번째 행동 선택
        if np.random.random() < current_epsilon:
            action = np.random.choice(env.n_actions)
        else:
            action = np.argmax(Q[state])
        
        done = False
        step_count = 0
        max_steps = 200
        episode_reward = 0
        
        while not done and step_count < max_steps:
            # 행동 실행
            next_state, reward, done = env.step(action)
            episode_reward += reward
            
            # 다음 행동 선택 (epsilon-greedy)
            if not done:
                if np.random.random() < current_epsilon:
                    next_action = np.random.choice(env.n_actions)
                else:
                    next_action = np.argmax(Q[next_state])
            else:
                next_action = 0  # 종료 상태에서는 의미 없음
            
            # SARSA 업데이트: Q(S,A) ← Q(S,A) + α[R + γQ(S',A') - Q(S,A)]
            old_q = Q[state][action]
            if not done:
                td_target = reward + gamma * Q[next_state][next_action]
            else:
                td_target = reward
            
            Q[state][action] += alpha * (td_target - Q[state][action])
            
            # 업데이트 기록 (선택적으로)
            if episode < 100 and episode % 10 == 0:
                q_updates.append({
                    'episode': episode,
                    'state': state,
                    'action': action,
                    'old_q': old_q,
                    'new_q': Q[state][action],
                    'td_error': td_target - old_q
                })
            
            # 다음 스텝으로
            state = next_state
            action = next_action
            step_count += 1
        
        episode_rewards.append(episode_reward)
        episode_lengths.append(step_count)
        
        # 주기적으로 진행상황 출력
        if (episode + 1) % 1000 == 0:
            recent_rewards = episode_rewards[-100:]
            recent_lengths = episode_lengths[-100:]
            avg_reward = np.mean(recent_rewards)
            avg_length = np.mean(recent_lengths)
            print(f"에피소드 {episode + 1}: 평균 보상 = {avg_reward:.2f}, 평균 길이 = {avg_length:.1f}, ε = {current_epsilon:.3f}")
    
    # 최종 그리디 정책 추출
    policy = {}
    for state in Q:
        policy[state] = np.argmax(Q[state])
    
    return policy, dict(Q), episode_rewards, episode_lengths, q_updates

# SARSA 실행
sarsa_policy, Q_sarsa, sarsa_rewards, sarsa_lengths, sarsa_updates = sarsa(
    env, num_episodes=6000, alpha=0.1, epsilon=0.1)

print(f"\nSARSA 학습 결과:")
print(f"학습된 상태-행동 쌍 수: {len(Q_sarsa)}")
```

```python
# 7-2. Q-Learning (개선된 버전)
print("\n7-2. Q-Learning Algorithm")
print("-" * 30)

def q_learning(env, num_episodes=3000, alpha=0.1, gamma=1.0, epsilon=0.1, epsilon_decay=0.995):
    """
    Q-Learning 알고리즘: Off-policy TD 제어 (개선된 버전)
    """
    Q = defaultdict(lambda: np.zeros(env.n_actions))
    episode_rewards = []
    episode_lengths = []
    q_updates = []
    
    print(f"Q-Learning 시작 ({num_episodes}개 에피소드)")
    print(f"학습률: {alpha}, 할인율: {gamma}, 초기 탐험률: {epsilon}")
    
    for episode in range(num_episodes):
        state = env.reset()
        done = False
        step_count = 0
        max_steps = 200
        episode_reward = 0
        
        # 현재 에피소드의 epsilon 값
        current_epsilon = epsilon * (epsilon_decay ** episode)
        
        while not done and step_count < max_steps:
            # epsilon-greedy로 행동 선택
            if np.random.random() < current_epsilon:
                action = np.random.choice(env.n_actions)
            else:
                action = np.argmax(Q[state])
            
            # 행동 실행
            next_state, reward, done = env.step(action)
            episode_reward += reward
            
            # Q-Learning 업데이트: Q(S,A) ← Q(S,A) + α[R + γ*max_a Q(S',a) - Q(S,A)]
            old_q = Q[state][action]
            if not done:
                td_target = reward + gamma * np.max(Q[next_state])
            else:
                td_target = reward
            
            Q[state][action] += alpha * (td_target - Q[state][action])
            
            # 업데이트 기록 (선택적으로)
            if episode < 100 and episode % 10 == 0:
                q_updates.append({
                    'episode': episode,
                    'state': state,
                    'action': action,
                    'old_q': old_q,
                    'new_q': Q[state][action],
                    'td_error': td_target - old_q
                })
            
            state = next_state
            step_count += 1
        
        episode_rewards.append(episode_reward)
        episode_lengths.append(step_count)
        
        # 주기적으로 진행상황 출력
        if (episode + 1) % 1000 == 0:
            recent_rewards = episode_rewards[-100:]
            recent_lengths = episode_lengths[-100:]
            avg_reward = np.mean(recent_rewards)
            avg_length = np.mean(recent_lengths)
            print(f"에피소드 {episode + 1}: 평균 보상 = {avg_reward:.2f}, 평균 길이 = {avg_length:.1f}, ε = {current_epsilon:.3f}")
    
    # 최종 그리디 정책 추출
    policy = {}
    for state in Q:
        policy[state] = np.argmax(Q[state])
    
    return policy, dict(Q), episode_rewards, episode_lengths, q_updates

# Q-Learning 실행
qlearning_policy, Q_qlearning, qlearning_rewards, qlearning_lengths, qlearning_updates = q_learning(
    env, num_episodes=6000, alpha=0.1, epsilon=0.1)

print(f"\nQ-Learning 학습 결과:")
print(f"학습된 상태-행동 쌍 수: {len(Q_qlearning)}")
```

- Q-Learning: 최적 정책 학습, 탐험 효율성

```python
# 정책 경로 분석
def analyze_policy_path(env, policy, start_state=None):
    """정책을 따라 경로 분석"""
    if start_state is None:
        start_state = env.start_state
    
    path = []
    state = start_state
    visited = set()
    max_steps = 50
    
    for step in range(max_steps):
        if state in visited or state == env.goal_state or state in env.cliff_states:
            break
        
        visited.add(state)
        path.append(state)
        
        if state in policy:
            action = policy[state]
            # 다음 상태 계산 (환경 실행 없이)
            row, col = env.state_to_coord(state)
            d_row, d_col = env.action_effects[action]
            new_row = max(0, min(env.rows - 1, row + d_row))
            new_col = max(0, min(env.cols - 1, col + d_col))
            state = env.coord_to_state(new_row, new_col)
        else:
            break
    
    return path

# 정책 경로 분석
print(f"\n🛤️ 학습된 정책 경로 분석:")
sarsa_path = analyze_policy_path(env, sarsa_policy)
qlearning_path = analyze_policy_path(env, qlearning_policy)

print(f"SARSA 경로 길이: {len(sarsa_path)} 스텝")
print(f"Q-Learning 경로 길이: {len(qlearning_path)} 스텝")
```

```python
# 경로 시각화
def visualize_policy_paths(env, sarsa_path, qlearning_path):
    """정책 경로 시각화"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    
    for ax, path, title, color in [(ax1, sarsa_path, "SARSA Policy Path", 'blue'),
                                   (ax2, qlearning_path, "Q-Learning Policy Path", 'red')]:
        # 기본 환경 그리기
        for i in range(env.rows + 1):
            ax.axhline(i - 0.5, color='black', linewidth=1)
        for j in range(env.cols + 1):
            ax.axvline(j - 0.5, color='black', linewidth=1)
        
        # 특수 상태들 표시
        for state in range(env.n_states):
            row, col = env.state_to_coord(state)
            display_row = env.rows - 1 - row
            
            if state in env.cliff_states:
                ax.add_patch(Rectangle((col-0.4, display_row-0.4), 0.8, 0.8, 
                                     facecolor='red', alpha=0.7))
                ax.text(col, display_row, 'C', ha='center', va='center',
                       fontsize=10, fontweight='bold', color='white')
            elif state == env.start_state:
                ax.add_patch(Rectangle((col-0.4, display_row-0.4), 0.8, 0.8,
                                     facecolor='green', alpha=0.7))
                ax.text(col, display_row, 'S', ha='center', va='center',
                       fontsize=14, fontweight='bold', color='white')
            elif state == env.goal_state:
                ax.add_patch(Rectangle((col-0.4, display_row-0.4), 0.8, 0.8,
                                     facecolor='blue', alpha=0.7))
                ax.text(col, display_row, 'G', ha='center', va='center',
                       fontsize=14, fontweight='bold', color='white')
        
        # 경로 그리기
        if len(path) > 1:
            for i in range(len(path) - 1):
                row1, col1 = env.state_to_coord(path[i])
                row2, col2 = env.state_to_coord(path[i + 1])
                
                display_row1 = env.rows - 1 - row1
                display_row2 = env.rows - 1 - row2
                
                ax.arrow(col1, display_row1, col2 - col1, display_row2 - display_row1,
                        head_width=0.1, head_length=0.1, fc=color, ec=color,
                        linewidth=3, alpha=0.8)
                
                # 스텝 번호 표시
                ax.text(col1, display_row1, str(i+1), ha='center', va='center',
                       fontsize=8, fontweight='bold', 
                       bbox=dict(boxstyle="circle,pad=0.1", facecolor='white', alpha=0.8))
        
        ax.set_xlim(-0.5, env.cols - 0.5)
        ax.set_ylim(-0.5, env.rows - 0.5)
        ax.set_aspect('equal')
        ax.set_title(title, fontsize=14, fontweight='bold')
        ax.set_xlabel('열 (Column)')
        ax.set_ylabel('행 (Row)')
    
    plt.tight_layout()
    plt.show()

visualize_policy_paths(env, sarsa_path, qlearning_path)
```

```python
# 추가 성능 메트릭
def calculate_safety_score(path, cliff_states):
    """안전성 점수 계산 (절벽에 인접한 상태의 비율)"""
    if not path:
        return 0
    
    adjacent_to_cliff = 0
    for state in path:
        row, col = env.state_to_coord(state)
        # 절벽 상태들과의 맨하탄 거리 계산
        min_cliff_distance = float('inf')
        for cliff_state in cliff_states:
            cliff_row, cliff_col = env.state_to_coord(cliff_state)
            distance = abs(row - cliff_row) + abs(col - cliff_col)
            min_cliff_distance = min(min_cliff_distance, distance)
        
        if min_cliff_distance <= 1:  # 절벽에 인접
            adjacent_to_cliff += 1
    
    return adjacent_to_cliff / len(path)

sarsa_safety = calculate_safety_score(sarsa_path, env.cliff_states)
qlearning_safety = calculate_safety_score(qlearning_path, env.cliff_states)

print(f"\n🛡️ 안전성 분석 (절벽 인접 비율):")
print(f"SARSA: {sarsa_safety:.2%}")
print(f"Q-Learning: {qlearning_safety:.2%}")
```

4. SARSA vs Q-Learning의 핵심 차이
   - On-policy vs Off-policy 학습 방식
   - 안전성 vs 최적성의 트레이드오프
   - 실제 응용에서의 선택 기준 이해

- 안전성과 효율성 간의 실제적 트레이드오프 확인

✓ 주요 실험 결과:
- 수렴 속도: TD 방법 > Monte Carlo
- 안전성: SARSA > Q-Learning
- 최적성: Q-Learning ≥ SARSA
- 구현 복잡도: TD(0) < SARSA ≈ Q-Learning < Monte Carlo
